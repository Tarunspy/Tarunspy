```python
#import basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Churn_Modelling.csv')
dataset
```

```python
#divide dataset independent and dependent features
#independent
X=dataset.iloc[:,3:13]
#dependent
y= dataset.iloc[:,13]
```

```python
X.head()
```

```python
y
```

```python
##feature engineering
geography=pd.get_dummies(X['Geography'],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
geography
```

```python
gender
```

```python
#concatenate these variables with dataframes
X=X.drop(['Geography','Gender'],axis=1)
```

```python
X.head()
```

```python
#concatenate these variables with dataframes
X=pd.concat([X,geography,gender],axis=1)
```

```python
#split dataset into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
#part 2 Now create ANN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU
from tensorflow.keras.layers import Dropout
```

```python
#lets initialize ANN
classifier= Sequential()
```

```python
X_train.shape
```

```python
classifier.add(Dense(units=11,activation='relu'))
```

```python
#adding first hidden layer
classifier.add(Dense(units=7,activation='relu'))
classifier.add(Dropout(0.2))
```

```python
#adding second hidden layer
classifier.add(Dense(units=6,activation='relu'))
```

```python
#adding output layer
classifier.add(Dense(1,activation='sigmoid'))
```

```python
classifier.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#adam will automatically provide learning rate =0.01 and if u want to initialize learning then
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
#early stopping when acuuracy is not increasing we will stop
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0
)
```

```python
model_history= classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=early_stopping )
```

```python
model_history.history.keys() #which parameters we focuse on that gives result
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#summarize hisory of accuracy
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
#part 3 - Making predictions and evaluating model
y_pred=classifier.predict(X_test)
y_pred=(y_pred >=0.5)
```

```python
## make confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
#caluculate accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
